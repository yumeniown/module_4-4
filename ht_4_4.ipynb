#### Задание

- Создать простую модельку на выбор(можно те, которые мы использовали ранее)
- Сериализовать pipeline в файл

```python
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.base import TransformerMixin, BaseEstimator
import numpy as np
```

Загружаем датасет об ирисах


```python
X, y = load_iris(return_X_y=True)
```

Создадим кастомный трансформер для добавления нового признака и встроим этот трансформер в сам пайплайн — для этого достаточно добавить новый шаг в пайплайн


```python
class Transformer(TransformerMixin, BaseEstimator):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        new_feature = X[:, 0] * X[:, 1]
        new_feature = new_feature.reshape(-1, 1)
        return np.hstack((X, new_feature))

pipe = Pipeline([
    ('FeatureEngineering', Transformer()),
    ('Scaling', StandardScaler()),
    ('FeatureSelection', SelectKBest(f_classif, k=3)),
    ('Regression', LogisticRegression()),
])

pipe.fit(X, y)
```

Пайплайн обучен. Давайте сохраним его в файл с помощью pickle:

```python
with open('iris_advanced_pipeline.pkl', 'wb') as file:
    pickle.dump(pipe, file)
```

Если сериализация завершилась успешно, то при инференсе модели мы сможем восстановить её из файла:

```python
with open('iris_advanced_pipeline.pkl', 'rb') as file:
    loaded_pipe = pickle.load(file)
```

Проверим, что результаты исходного и десериализованного пайплайнов и идентичны:



```python
print(all(pipe.predict(X) == loaded_pipe.predict(X)))
```
